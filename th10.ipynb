```python
def m(n):
    s = list(str(n))
    l = [-1] * 10
    for i, c in enumerate(s):
        l[int(c)] = i
    for i, c in enumerate(s):
        for d in range(int(c)):
            if l[d] > i:
                s[i], s[l[d]] = s[l[d]], s[i]
                return int(''.join(s))
    s[-1], s[-2] = s[-2], s[-1]
    return int(''.join(s))
n=input()
print(m(n))
#W9A3
```

```python
#W9A4
def can_transform(a, b):
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True
    for i in range(1, n + 1):
        for j in range(m + 1):
            if a[i - 1].islower():
                dp[i][j] |= dp[i - 1][j]
                if j > 0 and a[i - 1].upper() == b[j - 1]:
                    dp[i][j] |= dp[i - 1][j - 1]
            else:
                if j > 0 and a[i - 1] == b[j - 1]:
                    dp[i][j] |= dp[i - 1][j - 1]

    return "YES" if dp[n][m] else "NO"
Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    print(can_transform(a, b))
```

```python
#W7A7
def tc(a,k):
    c=0
    for i in range(len(a)):
        if a[i]==0:
            if i==0 and a[i+1]==0:
                a[1]==1
                c+=1
            elif i==len(a)-1 and a[i-1]==0:
                a[i-1]=1
                c+=1
            elif a[i]==0 and a[i+1]==0 and a[i-1]==0:
                a[i]=1
                c+=1
    return c>=k
b=list(map(int,input().split()))
n=int(input())
print(tc(b,n))

```

```python
#W9A8
def is_path_crossing(moves):
    bd=(0,0)
    a=list(moves)
    b=[]
    for i in a:
        b.append(bd)
        if i=='R':
            bd=(bd[0]+1,bd[1])
        elif i=='L':
            bd=(bd[0]-1,bd[1])
        elif i=='U':
            bd=(bd[0],bd[1]+1)
        elif i=='D':
            bd=(bd[0],bd[1]-1)
        if bd in b:
            return True
    return False
n=input()
print(is_path_crossing(n))

```
