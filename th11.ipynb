```python
try:
    a = int(input("Nhập a: "))
    if a <= 0:
        raise ValueError
    seen = set()
    while a != 1 and a not in seen:
        seen.add(a)
        s = 0
        while a > 0:
            n = a % 10
            s +=n ** 2
            a //= 10
        a = s
    if a == 1:
        print("Yes")
    else:
        print("No")
except ValueError:
    print("Lỗi nhập liệu: Vui lòng nhập số nguyên dương")


```

```python
t = int(input().strip())
for _ in range(t):
    s = input().strip()
    valid = True
    prev_repeat = False   
    i = 0
    while i < len(s):
        if s[i] == '\\':         
            if i == len(s) - 1:  
                valid = False
                break
            prev_repeat = False
            i += 2
            continue
        if s[i] in '*+':         
            if prev_repeat:      
                valid = False
                break
            prev_repeat = True
        else:
            prev_repeat = False
        i += 1
    print("True" if valid else "False") 
```
