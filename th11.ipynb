```python
try:
    a = int(input("Nhập a: "))
    if a <= 0:
        raise ValueError
    seen = set()
    while a != 1 and a not in seen:
        seen.add(a)
        s = 0
        while a > 0:
            n = a % 10
            s +=n ** 2
            a //= 10
        a = s
    if a == 1:
        print("Yes")
    else:
        print("No")
except ValueError:
    print("Lỗi nhập liệu: Vui lòng nhập số nguyên dương")


```
