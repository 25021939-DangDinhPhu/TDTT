```python
def reverse(path: str) -> int:
    with open(path, "r") as f:
        n = f.read().strip()       
    a = n[::-1]       
    return int(a)

```

```python
def t(path: str):
    count = 1
    with open(path, 'r') as f:
        l = f.readlines()
        a = list(map(int, l[1].split()))
    for _ in range(3):
        n = max(a)
        count *= n
        a.remove(n)
    return count
```

```python
def t(path: str):
    with open(path, 'r') as f:
        l = f.readlines()
        a = list(map(int, l[1].split()))
    while 0 in a:
        a.remove(0)
        a.append(0)
    return a
```

```python
def findMovies(path: str) -> list[str]:
    r = []
    with open(path, 'r') as f:
        li = f.readlines()

    for i in li[1:]:  # bá» header
        parts = [p.strip() for p in i.split(',')]
        movie = parts[1]
        description = parts[2].lower()
        rating = float(parts[3])
        if rating > 8.0 and "boring" not in description:
            r.append(movie)
    return r
```

```python
import math
def printPerimeters(path: str):
    with open(path, 'r') as f:
        l = [line.strip() for line in f if line.strip()]
    i = 0
    while i < len(l):
        shape = l[i].split(':')[1].strip()
        data = l[i+1]
        if shape == "SQUARE":
            side = float(data.split(':')[1])
            p = 4 * side
        elif shape == "RECTANGLE":
            parts = data.split()
            w = float(parts[1])
            h = float(parts[3])
            p = 2 * (w + h)
        elif shape == "CIRCLE":
            r = float(data.split(':')[1])
            p = 2 * math.pi * r
        print(f"{p:.2f}")
        i += 2
```

```python
import math
def estimatedTime(path: str) -> float:
    with open(path, 'r') as f:
        l = f.readlines()
    x1, y1 = map(float, l[0].split(':')[1].split())
    x2, y2 = map(float, l[1].split(':')[1].split())
    v = float(l[2].split(':')[1])
    d = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return f"{d/v:.2f}"
```

```python
import struct
def readDoubles(path: str):
    with open(path, 'rb') as f:
        while True:
            b = f.read(8)    
            if not b:
                break
            v = struct.unpack('d', b)[0]
            print(f"{v:.5f}")
```

```python
def imageSmoother(path: str):
    with open(path, 'r') as f:
        M = int(f.readline().strip())
        img = [list(map(int, f.readline().split())) for _ in range(M)]
    r= [[0]*M for _ in range(M)]
    for i in range(M):
        for j in range(M):
            s = 0
            cnt = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    x = i + dx
                    y = j + dy
                    if 0 <= x < M and 0 <= y < M:
                        s += img[x][y]
                        cnt += 1
            r[i][j] = s // cnt 
    for row in r:
        print(*row)
```
