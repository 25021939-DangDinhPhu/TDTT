```python
a = list(map(int, input().split()))
k = int(input())
res = []
for i in range(0,len(a)):
    for j in range(i + 1,len(a)):
        if a[i] + a[j] == k:
            if (a[i],a[i]) not in res:
                res.append((a[i], a[j]))
res.sort(key=lambda x: x[0])
print(res)



```

```python
def kt(m,n):
    for i in range(len(m)):
        if m[i]==n:
            return i+1
    return -1
a=list(map(int,input().split()))
k=int(input())
print(kt(a,k))

```
