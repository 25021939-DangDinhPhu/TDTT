```python
def d(b):
    n = len(b)
    dp = [1] * n
    for i in range(n):
        for j in range(i):
            if b[j] < b[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp) if b else 0
a=list(map(int,input().split()))
print(d(a))
```
