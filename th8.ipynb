```python
class ht:
    def __init__(self,h,r):
        self.h=h
        self.r=r
    def dientich(self):
        return 2*3.14*self.h*self.r+2*3.14*(self.r)**2  
    def thetich(self):
        return 3.14*(self.r)**2*self.h
a,b=map(float,input().split())
c=ht(a,b)
print(f'{c.dientich():.2f}',f'{c.thetich():.2f}')
```

```python
class  Rectangle:
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return (self.w*+self.h)
    def perimeter(self):
        return 2*(self.w+self.h)
    def scale(self,k):
        self.w=self.w*k
        self.h=self.h*k
x,y,k=map(float,input().split())
a=Rectangle(x,y)
print(round(a.area(),2),round(a.perimeter(),2))
```

```python
from math import*
class fuction:
    def __init__(self,num,den):
        if den<0:
            num=-num
            den=-den
        g=gcd(num,den)
        self.num=num//g
        self.den=den//g
    def add(self,fution2):
        self.num=self.num*fution2.den+self.den*fution2.num
        self.den=self.den*fution2.den
    def sub(self,fution2):
        self.num=self.num*fution2.den-self.den*fution2.num
        self.den=self.den*fution2.den
    def mul(self,fution2):
        self.num=self.num*fution2.num
        self.den=self.den*fution2.den
    def div(self,fution2):
        if fution2.num!=0:
            self.num=self.num*fution2.den
            self.den=self.den*fution2.num
        else:
            print('không thể chia')
    def __str__(self):
        return f"{self.num}/{self.den}"
a,b,op,d,e=input().split()
a,b,d,e=int(a),int(b),int(d),int(e)
if b==0 or d==0:
    print('phân số không tồn tại')
else:
    f1=fuction(a,b)
    f2=fuction(d,e)
    if op=='+':
        f1.add(f2)
    elif op=='-':
        f1.sub(f2)
    elif op=='*':
        f1.mul(f2)
    elif op=='/':
        f1.div(f2)
    f3=fuction(f1.num,f1.den)
    print(f3)
        

```

```python
class Student():
    def __init__(self,sub,score):
        self.sub=sub
        self.score=score
    def avg(self,m):
        a=2
        self.score=(self.score+m.score)/a
        a+=1
    def rank(self):
        if self.score>=8:
            return 'Excellent'
        elif 6.5<=self.score<8:
            return 'Good'
        elif 5<=self.score<6.5:
            return 'average'
        else:
            return 'poor'
ten=input()
n=int(input())
for i in range(n):
    mon,diem=input().split()
    if i==0:
        m1=Student(mon,float(diem))
    else:
        m2=Student(mon,float(diem))
        m1.avg(m2)
print(ten,f'{m1.score:.2f}',m1.rank())

```

```python
from math import*
class d:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def vth(self,a):
        return v(a.x-self.x,a.y-self.y,a.z-self.z)
class v:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def tch(self,b):
        return v(
            self.y * b.z - self.z * b.y,
            self.z * b.x - self.x * b.z,
            self.x * b.y - self.y * b.x
        )
    def tvh(self,b):
        return self.x*b.x+self.y*b.y+self.z*b.z
    def do_dai(self):
        return (self.tvh(self))**0.5
class mp:
    def __init__(self,p1,p2,p3):
        v1=p1.vth(p2)
        v2=p1.vth(p3)
        self.g=v1.tch(v2)
A = d(*map(float, input().split()))
B = d(*map(float, input().split()))
C = d(*map(float, input().split()))
D = d(*map(float, input().split()))
mp1=mp(A,B,C)
mp2=mp(B,C,D)
n1=mp1.g
n2=mp2.g
cos_alpha=abs(n1.tvh(n2)/(n1.do_dai()*n2.do_dai()))
alpha=degrees(acos(cos_alpha))
print(round(alpha,2))

        
```

```python

class Triangle:
    def __init__(self,x,y,z):
        self.x =x
        self.y =y
        self.z =z
    def getArea(self):
        a, b, c = self.x, self.y, self.z
        if a <= 0 or b <= 0 or c <= 0:
            return -1
        if a + b <= c or a + c <= b or b + c <= a:
            return -1
        p = (a + b + c) / 2
        return ((p * (p - a) * (p - b) * (p - c)))**0.5
a, b, c = map(int, input().split())
triangle = Triangle(a, b, c)
area = triangle.getArea()
if area == -1:
    print("invalid")
else:
    print(f"{area:.2f}")

        
```
